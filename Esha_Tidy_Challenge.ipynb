```python
# Import necessary libraries
import json
import pandas as pd


# Load the JSON file as a list of dictionaries
with open ("c:\\Users\\Esha\\Inderactive Data_\\chessbuds_messages.json") as f:
    messages = json.load(f)
```

### The json file contains the following keys.


Participants, Messages, Title, Is_still_participant, Thread_type, Thread_path, Magic_words, Joinable_mode

The provided Json file can be devided into two major componentnts listed below:

- Participants contains list of people involved in the conversation.
- Messages contains list of messages exchanged in the conversation.

Considering the above components we can say that "Messages" can be best suited to create a data frame for further manipulation.

```python
df = pd.json_normalize(messages)
df.head()
```

```python
# Now creating a DataFrame 
df_2 = pd.json_normalize(
    messages, record_path='messages'
)
# Drop rows containing NaN values
df_cleaned = df_2.dropna()
df_2
```

```python
df_2.columns
```

```python
# Dropping columns
columns_to_drop = ['reactions',
       'is_unsent', 'is_taken_down', 'bumped_message_metadata.bumped_message',
       'bumped_message_metadata.is_bumped', 'share.link', 'photos', 'gifs',
       'users']

df_drop = df_2.drop(columns=columns_to_drop)
# Drop duplicates based on the 'content' column
df_deduplicated = df_drop.drop_duplicates(subset='content', keep='first')
df_drop
```
