```python
# import libraries 
import pandas as pd
import matplotlib.pyplot as mlt

# import the csv file and convert it to a dataframe object

df = pd.read_csv('c:\\Users\\Esha\\Inderactive Data_\\cces_example_data (1).csv')

df
```

```python
# Create an empty figure with two axes
fig, axes= mlt.subplots(1,2, figsize=(10,4))
axes[0].set_title("Age histogram")
axes[1].set_title("Political identity")
```

```python
# Barplot that shows frequencies of responses in evaluation variables 
# i have used the same axes as above and changes its titles to suit the current visual being viewed below
fig2, axes[0] = mlt.subplots()

df['Evaluation'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='pink')

# Setting axes lables and titles 
axes[0].set_title('Evaluation response frequency')
axes[0].set_ylabel('response')
axes[0].set_xlabel('Evaluation')
```

```python
# creating a scatterplot  on `ConMediaUse` Vs `ImmiPhobia`

fig3, axes4 = mlt.subplots()

df.plot.scatter(x='ConMediaUse',y='ImmiPhobia',ax=axes4,s=40,color = 'red',alpha=0.5)

# Setting axes lables and titles 
axes4.set_title('`ConMediaUse` Vs `ImmiPhobia`')
axes4.set_ylabel('ImmiPhobia')
axes4.set_xlabel('ConMediaUse')
```

### How to improve the above Scattrplot?

Scatterplots are a great way of visualizing data when we especially have two continuous quantitative variables. And that is waht we have done above by postioning  'ConMediaUse' on x axis and 'Immiphobia' on the y axis.

Scatter plots are likely to overlap irrespective of the size of the data being used. For example it might happen in smaller datasets aswell if 'Data values are rounded with low precision', such that multiple observations have exactly theh same value. 

We can choose to represent the position the data value of X axis using a shape scale and adjust the transpiracy levels in colour to differentiate. Along with that we can use of data lables to identify key data points though it might make the sctter plot look more busier. it helps us gain some clarity. Transparency in colour might get the job done yet jittring might move it to the next level, it helps data points clearer and visable. However jittering is to be used with caution as it might lead to an inaacurate representation.
